```python
naive_times = pd.DatetimeIndex(start='2016', end='2017', freq='1h') #creating datetimes for 1y with 1h freq
sandia_modules = pvlib.pvsystem.retrieve_sam('SandiaMod')           #retrieving solar panels list
sapm_inverters = pvlib.pvsystem.retrieve_sam('cecinverter')         #retrieving inverters list
module = sandia_modules['Canadian_Solar_CS5P_220M___2009_']         #add the preferable solar panel
#inverter = sapm_inverters['ABB__MICRO_0_25_I_OUTD_US_208_208V__CEC_2014_'] 
cec_inverter = sapm_inverters['ABB__MICRO_0_25_I_OUTD_US_208_208V__CEC_2014_'] #add the preferable inverter
times = naive_times.tz_localize('Etc/GMT-1')
location = Location(52.8, 4.9, name='Amsterdam', altitude=-2, tz='Etc/GMT-1')
```

```python
opt_annual_energy = 0         #starting from 0 to find optimal values for tilt, azimuth and max energy 
opt_tilt = 0                
opt_azimuth = 0

for i in range(90):
    for j in range(360):
        system = PVSystem(surface_tilt = i, surface_azimuth = j, module_parameters = module, 
                          inverter_parameters = cec_inverter)  #create system with custom parameters
        mc = ModelChain(system, location)                      #creating the relation between the system and the location
        import pdb; pdb.set_trace()
        mc.run_model(times)                                    #running the simulation for 1 year, 1h intervals
        annual_energy = mc.ac.sum()                            #calculating annual yield
        print('Tilt: ' , i , '\n', 'Azimuth: ', j ,'\n', 'Annual Energy: ' , annual_energy)    
        if annual_energy > opt_annual_energy:                  #calculating the max yield
            opt_annual_energy = annual_energy
            opt_tilt = i
            opt_azimuth = j
            print('Optimal Tilt: ' , opt_tilt , '\n', 'Optimal Azimuth: ' , opt_azimuth , '\n', 'Optimal Annual Energy: ' , opt_annual_energy)
        

print('Optimals:\n' , '\t' , opt_annual_energy , '\n\t' , opt_tilt)
```
